```python
from sympy.mpmath import *
mp.dps = 25; mp.pretty = True
%matplotlib inline
from sympy import *
init_printing()
theta, psidot, chi = symbols(r'\theta \dot{\psi} \chi')
t = symbols('t')
```

```python
inner = (((cos(theta) -1)*sin(psidot * t) * sin(chi) - sin(theta) * cos(chi))/
          (cos(psidot * t)*sin(chi) + (cos(theta) * sin(psidot * t) * sin(chi) - sin(theta) * cos(chi)) * tan(psidot)))
inner
```

```python
DeltaPhi = atan(inner)
DeltaPhi
```

```python
diff(atan(inner), t)
```

```python
DeltaPhi2 = diff(atan(inner), t, t)
DeltaPhi2
```

```python
plot(DeltaPhi2.subs({chi: 80, theta:3.0, psidot:0.1}), (t, -100, 100), nb_of_points=10000)
```
