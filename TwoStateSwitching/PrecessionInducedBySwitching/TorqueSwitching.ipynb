```python
def PLOT(epsI1=0.0, epsI3=3.0e-3, epsA=5.0e-4 , omega0=10,
         error=1e-12, T=3.0e3 , chi0 = 70.0, AnomTorque=True ,
         a0=20.0, upsilon=0.0, n=20000, cleanup=True,
         order=3, divisor=10):
    
    file_name = main(epsI1, epsI3, epsA , omega0, chi0, a0, T, 
                     AnomTorque, upsilon, n, error, cleanup)

    ax1 = Plot.PhaseResidual(file_name, order=order)

    ax2 = Plot.SpindownRate(file_name, divisor=divisor)

    plt.show()
```
