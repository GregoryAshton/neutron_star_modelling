```python
import matplotlib.pylab as plt
import numpy as np
import scipy as sc
from scipy.integrate import cumtrapz
from nsmod import Useful_Tools

%matplotlib inline
import TNtools
TNtools.PlotDefaults()

import TempoTools as TT
import PhaseResidualsTools as PRT

import scipy.signal
```

```python
# Simulation Paramaters
F1A = -7.4e-14
F1B = -5.0e-14
D = 1e-10
T = 100 * 86400
R = 0.6
tA = R * T
tB = T - tA
F0_init = 2.0
N = 1000 # Number of flips
Tobs = 20 * T

tA_list = np.random.normal(tA, D*tA, size=N)
tB_list = np.random.normal(tB, D*tB, size=N)

flip_markers = [sum(np.dstack((tA_list,tB_list)).flat[0:i]) for i in range(2*N)]

time = np.linspace(0, Tobs, 10000)

F1_list=[]
current_F1, other_F1 = F1A, F1B
j=0
for t in time:
    if t < flip_markers[j]:
        F1_list.append(current_F1)
    if t >= flip_markers[j]:
        current_F1, other_F1 = other_F1, current_F1
        j+=1
        F1_list.append(current_F1)

# Plotting 
fig, (ax1, ax2) = plt.subplots(nrows=2)

ax1.plot(time, F1_list, "-")
ax1.set_ylabel(r"Spin-down rate", rotation="vertical")
ax1.set_xticklabels([])
ax1.set_yticks([F1A, F1B])
ax1.set_yticklabels([r"$\dot{\nu}_{A}$", r"$\dot{\nu}_{B}$"])
ax1.set_ylim(1.1*min(F1B, F1A), 0.9*max(F1A, F1B))
  
    
F0_list = 2 + cumtrapz(y=F1_list, x=time, initial=0)
phase = 2 * np.pi * cumtrapz(y=F0_list, x=time, initial=0)

         
# Fit polynomial to phase of order order
coefs = np.polyfit(time, phase, 2)
# poly1d returns the polynomial we then evaluate this at time giving the fitted phi
phase_fit = np.poly1d(coefs)(time)

# Subtract the two to get a residual
phase_res = phase - phase_fit

ax2.plot(time, phase_res, color="k", label="Numerical")
ax2.set_ylabel(r"Phase residual")
ax2.set_xlabel(r"Time [s]")
ax2.set_yticks(ax2.get_yticks()[0:-1])
#ax2.text(0.9*time[-1], 0.9 * ax2.get_xlim()[-1], "R={} \n D={}".format(R, D), size=12, 
#         bbox=dict(facecolor='white', edgecolor="white", alpha=0.9))

# Analytic calculation
dFdot_total = abs(F1A - F1B)
def DeltaPhi(t, dFdot_total, T, R):
    t = np.mod(t, T)
    if t < R * T:
        x1 = (1-R) * t * (t - R*T)
    if t >= R*T:
        x1 = R * (T - t) * (t - R*T)
        
    return np.pi * dFdot_total * x1 

DeltaPhi_list = [DeltaPhi(t, dFdot_total, T, R) for t in time]
ax2.plot(time, DeltaPhi_list, ls="--", label="Analytic")
leg = ax2.legend(framealpha=1.0, fontsize=12)

plt.subplots_adjust(hspace=0.0, wspace=0.5)
plt.savefig("img/verification_example.pdf".format(R, D))
plt.show()
```
