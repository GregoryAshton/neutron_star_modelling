```python
import matplotlib.pylab as plt
import numpy as np
import scipy as sc
from scipy.integrate import cumtrapz
from nsmod import Plot, Useful_Tools

%matplotlib inline
import TNtools
TNtools.PlotDefaults()

import TempoTools as TT
import PhaseResidualsTools as PRT
```

```python
# Data from paper
F0 = 2.469
F1 = -365.68e-17
DeltaF1_F1 = 0.71
F = 0.73

# Derived quantities
DeltaF1 = DeltaF1_F1 * F1
T = 86400 * 365 * (1./F)

# Made up quantities
R = 0.5

# Simulation parameters
tA = R * T
tB = (1-R) * T
multiplyer = 0.0065
F1A = F1 - multiplyer * .5 * DeltaF1 
F1B = F1 + multiplyer * .5 * DeltaF1 
F0_init = F0

# Time parameters

dT = 10 * 86400
Tobs = 4000*86400

# Generate the Switching
time, F1 = PRT.get_F1_LYNE(Tobs=Tobs, tA=tA, tB=tB, F1A=F1A, F1B=F1B, D=0.0, N=10000)
P0, F0 = PRT.GeneratePhase(time, F1, F0_init)

# Generate the residuals using tempo
time_Tempo, residuals_Tempo = TT.GetResiduals(dT, time, P0, F0, F1)

# Plotting

fig, (ax1, ax2) = plt.subplots(nrows=2)
ax1.plot(time, F1)
ax1.set_ylabel(r"Spin-down rate", rotation="vertical")
ax1.set_xticklabels([])
#ax1.set_yticks([F1A, F1B])
#ax1.set_yticklabels([r"$\dot{\nu}_{A}$", r"$\dot{\nu}_{B}$"])
ax1.set_ylim(1.1*min(F1B, F1A), 0.9*max(F1A, F1B))


ax2.plot(time_Tempo, residuals_Tempo, "k", label="Tempo method")
ax2.set_ylabel(r"Timing residual")
ax2.set_xlabel(r"Time [s]")
ax2.set_yticks(ax2.get_yticks()[0:-1])

plt.show()
```
