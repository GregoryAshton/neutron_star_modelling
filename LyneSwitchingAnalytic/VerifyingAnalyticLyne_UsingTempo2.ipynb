## Verifying the Lyne model using Tempo2

```python
# Simulation parameters
T0 = 1087212928
dT = 0.1*86400
tA = 40*86400.
tB = 10*86400.
F1A = -1.0e-15
F1B = -1.9e-15
F0_init = 1e-3
Tobs = 300*86400

# Generate the Switching
time, F1 = PRT.get_F1_LYNE(Tobs=Tobs, tA=tA, tB=tB, F1A=F1A, F1B=F1B, D=0.0, N=10000)
P0, F0 = PRT.GeneratePhase(time, F1, F0_init)

# Generate the residuals using tempo
time_Tempo, residuals_Tempo = TT.GetResiduals(dT, time, P0, F0, F1)

# Plotting

fig, (ax1, ax2) = plt.subplots(nrows=2)
ax1.plot(time, F1)
ax1.set_ylabel(r"Spin-down rate", rotation="vertical")
ax1.set_xticklabels([])
ax1.set_yticks([F1A, F1B])
ax1.set_yticklabels([r"$\dot{\nu}_{A}$", r"$\dot{\nu}_{B}$"])
ax1.set_ylim(1.1*min(F1B, F1A), 0.9*max(F1A, F1B))


ax2.plot(time_Tempo, residuals_Tempo, "g", label="Tempo method")


# Analytic calculation
dFdot_total = abs(F1A - F1B)
def DeltaPhi(t, dFdot_total, T, R):
    t = np.mod(t, T)
    if t < R * T:
        x1 = (1-R) * t * (t - R*T)
    if t >= R*T:
        x1 = R * (T - t) * (t - R*T)
        
    return -1 * np.pi * dFdot_total * x1 
T = tA + tB
R = tA/T
DeltaPhi_list = [DeltaPhi(t, dFdot_total, T, R) for t in time]

ax3 = ax2.twinx()
ax3.plot(time, DeltaPhi_list, ls="--", label="Analytic")

ax2.set_ylabel(r"Phase residual")
ax2.set_xlabel(r"Time [s]")
ax2.set_yticks(ax2.get_yticks()[0:-1])

plt.show()
```
