```python
from sympy import *
%matplotlib inline
from sympy import *
init_printing()
theta, chi = symbols(r'\theta \chi')
psi = Function('psi')
phi = Function('phi')
t = symbols('t')
```

# Spindown with $\theta=$const.

```python
Phi = atan(cos(psi(t)) * tan(chi)/(sin(theta) - sin(psi(t)) * cos(theta) * tan(chi)))
Phi
```

```python
Phi_dot = diff(Phi, t)
Phi_dot.trigsimp()
```

```python
Phi_ddot = diff(Phi_dot, t)
Phi_ddot.trigsimp()
```

```python
Phi_ddot.simplify()
```

```python
inner = (((cos(theta) -1)*sin(psidot * t) * sin(chi) - sin(theta) * cos(chi))/
          (cos(psidot * t)*sin(chi) + (cos(theta) * sin(psidot * t) * sin(chi) - sin(theta) * cos(chi)) * tan(psidot)))
inner
```

```python
DeltaPhi = atan(inner)
DeltaPhi
```

```python
diff(atan(inner), t)
```

```python
DeltaPhi2 = diff(atan(inner), t, t)
DeltaPhi2
```

```python
plot(DeltaPhi2.subs({chi: 80, theta:3.0, psidot:0.1}), (t, -100, 100), nb_of_points=10000)
```
