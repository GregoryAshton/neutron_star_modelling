```python
import numpy as np
from matplotlib import pyplot as plt
import nsmod
import os 
%matplotlib inline

from nsmod.manual_switching_torque_with_Euler import main
from nsmod import Plot, File_Functions, Physics_Functions
```

```python
epsI1 = -9e-6
epsI3 = 8e-6
omega0 = 1e2
epsA = 7e-9
chi0 = 71.8
tauP = 2 * np.pi/(epsI3 * omega0)
T = 6.2 * tauP
a0 = 2.0

n = 50000
error = 1e-13

file_name = main(chi0=chi0, epsI1=epsI1, epsI3=epsI3, epsA=epsA, omega0=omega0, T=T, 
                 n=n, error=error, a0=a0, cleanup=False, DryRun=False, 
                 AnomTorque=True
                 )

#File_Functions.PrintParameterDictionary(file_name)

fig, (ax1, ax2, ax3) = plt.subplots(nrows = 3, sharex=True, figsize=(8, 10))


ax1 = Plot.PhaseResidual(file_name, ax=ax1)
ax2 = Plot.SpindownRate(file_name, ax=ax2, analytic=["EM"], divisor=4)
ax3 = Plot.PulseWidth(file_name, Phi0=0.5, Theta0=np.pi, sigmaPhi=0.01, sigmaTheta=0.01, ax=ax3)

for ax in [ax1, ax2]: ax.set_xlabel(''); 
    
fig.subplots_adjust(hspace=0)

plt.savefig("Triaxial3.pdf")
plt.show()
```
